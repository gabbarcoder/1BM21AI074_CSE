```python
import numpy as np 
import pandas as pd 
import matplotlib.pyplot as plt 
%matplotlib inline 
import sklearn
```

```python
X=np.array(10*np.random.randn(37,5))
```

```python
df = pd.DataFrame(X, columns=['Feature'+str(l) for l in range (1,6)])
```

```python
df.head()
```

```python
for i in df.columns:
    df.hist(i,bins=5,xlabelsize=15,ylabelsize=15,figsize=(8,6))
```

```python
df['y']=5*df['Feature1']**2+13*df['Feature2' ]+0.1*df['Feature3']**2*df['Feature1']\
+2*df['Feature4']*df['Feature5']+0.1*df['Feature5']**3+0.8*df['Feature1']*df['Feature4']*df['Feature5'] \
+30*np.random.normal(loc=5,scale=2)
```

```python
df.head()
```

```python
for i in df.columns:
    df.plot.scatter(i,'y',edgecolors=(0,0,0),s=50,c='b')
```

```python
from sklearn.linear_model import LinearRegression
```

```python
l_m = LinearRegression()
```

```python
x = df.drop('y',axis=1)
```

```python
y=df['y']
```

```python
l_m.fit(x,y)
```

```python
y_p= l_m.predict(x)
```

```python
rmse=np.sqrt(np.sum(np.square(y_p-y)))
```

```python
print(rmse)
```

```python
coeff_linear = pd.DataFrame(l_m.coef_,index=df.drop('y',axis=1).columns,columns=['Linear Model Coefficient'])
coeff_linear
```
